```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv(r"C:\Users\s8908\Churn_Modelling.csv")
```

```python
dataset
```

```python
x=dataset.iloc[:,3:13]#1st column - include all
#2nd column - which columns
y=dataset.iloc[:,13]
```

```python
x
```

```python
y
```

```python
#create dummy variable
Geography = pd.get_dummies(x['Geography'],drop_first = True)
Gender    = pd.get_dummies(x['Gender'],drop_first=True)
```

```python
Gender,Geography
```

```python
x=x.drop(['Geography','Gender'], axis=1)
```

```python
x=pd.concat([x,Gender,Geography], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#FEATURE SCALING
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
import keras
from tensorflow.keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
#Initalizing ANN 
classifier = Sequential()
#Adding first input layer and first hidden layer 
classifier.add(Dense(units=6, kernel_initializer='he_uniform', activation='relu', input_dim=11))
```

```python
#Adding 2nd hidden layer
classifier.add(Dense(units=10, kernel_initializer='he_uniform', activation='relu'))
```

```python
#Adding output layer
classifier.add(Dense(units=1, kernel_initializer='glorot_uniform', activation='sigmoid'))
```

```python
#compiling Ann model
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#Fitting ANN to training set 
model_history = classifier.fit(X_train, y_train, validation_split=0.33, batch_size=10, epochs=100)
```

```python
print(model_history.history.keys())
```

```python
#On testing data
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
import numpy as np

# Assuming you have the predicted probabilities in the 'predictions' variable
predicted_probabilities = y_pred

# Convert probabilities to binary predictions
threshold = 0.5
predicted_classes = (predicted_probabilities >= threshold).astype(int)

print(predicted_classes)
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predicted_classes)
#calculate accuracy
from sklearn.metrics import accuracy_score
score = accuracy_score(predicted_classes,y_test)
```
